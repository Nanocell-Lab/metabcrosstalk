```python
def geneDF4cyc(codecyc):
    """get the genes fron cyc pathway"""
    import requests as rq
    import pandas as pd    
    #url = 'https://cyc.pnnl.gov/META/pathway-genes?object={0}'.format(codecyc)
    url = 'https://metacyc.org/META/pathway-genes?object={0}'.format(codecyc)
    handle = rq.get(url)
    text = handle.text
    data = text.split("\n")
    print(data[0])
    rows = [x.split("\t") for x in data[3:-1]]
    #print(rows)
    columns = data[2].split('\t')
    #print(columns)
    df = pd.DataFrame(rows, columns=columns)
    df.insert(0, "Pathways", data[0], True)
    return(df)
```

```python
def getpathways(word, cyc):
    """get pathway codes"""
    import requests as rq
    import pandas as pd 
    import re
    from bs4 import BeautifulSoup
    
    if cyc=='eco':
        url = 'https://ecocyc.org/ECOLI/substring-search?type=NIL&object={0}'.format(word)
    elif cyc=='hsa':
        url = 'https://humancyc.org/HUMAN/substring-search?type=NIL&object={0}'.format(word)
    elif cyc=='meta':
        url = 'https://metacyc.org/META/substring-search?type=NIL&object={0}'.format(word)
    elif cyc=='psa':
        url='https://pseudomonas.biocyc.org/PAER208964/substring-search?type=NIL&object={0}'.format(word)
    else:
        print("use default cyc MetaCyc...")
        url = 'https://metacyc.org/META/substring-search?type=NIL&object={0}'.format(word)
    
    handle = rq.get(url)
    text = handle.text
    sopa = BeautifulSoup(text)
    pathways_codes = []
    for link in sopa.find_all('a', attrs={'href': re.compile("PATHWAY&object=")}): 
        # display the actual urls 
        pathways_codes.append(link.get('href').split('=')[-1])

    return pathways_codes
```

```python
pathways_codes = getpathways(word_pathway, cyc=cyc)
print(pathways_codes)
```

```python
import pandas as pd
df_master = pd.DataFrame()
for pathway_code in pathways_codes:
    df_pathway = geneDF4cyc(pathway_code)
    df_master = pd.concat([df_master, df_pathway])
df_master.head(3)
```

```python
import altair as alt
bars = alt.Chart(df_master).mark_bar().encode(
    x=alt.X('count(Gene ID):Q'),
    y=alt.Y('Pathways:N')
#    color='year:N',
).configure_range(
    category={'scheme': 'accent'}
)
bars
```

```python
# Save the DB search
newfile = "Pathways_{0}_{1}.csv".format(cyc, word_pathway)
df_master.to_csv(newfile)
print(newfile)
```
